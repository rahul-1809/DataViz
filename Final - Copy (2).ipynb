```python
import pandas as pd
data=pd.read_csv("superstore_new.csv")
data
```

```python
data.shape
```

```python
data.drop(["Country","Ship.Date","Row.ID","Customer.Name","Market"],axis=1,inplace=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
data["Category"]=label_encoder.fit_transform(data["Category"])
data["City"]=label_encoder.fit_transform(data["City"])
data["Customer.ID"]=label_encoder.fit_transform(data["Customer.ID"])
data["Order.Priority"]=label_encoder.fit_transform(data["Order.Priority"])
data["Product.Name"]=label_encoder.fit_transform(data["Product.Name"])
data["Region"]=label_encoder.fit_transform(data["Region"])
data["Segment"]=label_encoder.fit_transform(data["Segment"])
data["Ship.Mode"]=label_encoder.fit_transform(data["Ship.Mode"])
data["State"]=label_encoder.fit_transform(data["State"])
data["Sub.Category"]=label_encoder.fit_transform(data["Sub.Category"])
data["Market2"]=label_encoder.fit_transform(data["Market2"])
data["Order.ID"]=label_encoder.fit_transform(data["Order.ID"])
data["Product.ID"]=label_encoder.fit_transform(data["Product.ID"])
data
```

```python
data.isnull().sum()
```

```python
data.fillna(data.mean(numeric_only=True), inplace=True)
data.head()
```

```python
data['OrderDate'] = pd.to_datetime(data['Order.Date'])
data['Year'] = data['OrderDate'].dt.year
data['Month'] = data['OrderDate'].dt.month
```

```python
data['OrderDate'] = pd.to_datetime(data['Order.Date'])
data['Year'] = data['OrderDate'].dt.year
data.drop(['Order.Date'], axis=1, inplace=True)

grouped_data = data.groupby('Year').mean()
grouped_data
```

```python
data.to_csv("preprocessed_dataset1.csv",index=False)
```
